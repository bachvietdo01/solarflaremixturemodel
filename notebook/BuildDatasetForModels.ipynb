```python
data = pd.read_csv('../fulldata/allflares6h.csv')
```

```python
X = data.iloc[:,:20].to_numpy()
y = data.iloc[:,20].to_numpy()
```

```python
def convert_to_cat(log_y):
    n = len(log_y)
    z = np.zeros(n)
    
    for i in range(n):
        if log_y[i] < -6 * np.log(10):
            z[i] = 1
        elif log_y[i] < -5 * np.log(10):
            z[i] = 2
        elif log_y[i] < -4 * np.log(10):
            z[i] = 3
        else:
            z[i] =4
    
    return z

def convert_to_bin_cat(log_y):
    n = len(log_y)
    z = np.zeros(n)
    
    for i in range(n):
        if log_y[i] < -5 * np.log(10):
            z[i] = 1
        else:
            z[i] = 2
    
    return z
```

```python
z = convert_to_bin_cat(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, z, test_size=0.35, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 100)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

print(f1_score(y_test, y_pred))
print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
```
